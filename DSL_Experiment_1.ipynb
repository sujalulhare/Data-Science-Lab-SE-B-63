```python
knowledge_base = {
    ('hot', 'caffeinated'): 'Coffee',
    ('hot', 'non-caffeinated'): 'Herbal Tea',
    ('cold', 'caffeinated'): 'Iced Coffee',
    ('cold', 'non-caffeinated'): 'Juice'
}

print("Knowledge Base for Drink Recommendations:")
for (temp, caffeine), drink in knowledge_base.items():
    print(f"  Temperature: {temp}, Caffeine: {caffeine} -> Recommended Drink: {drink}")
```

```python
# Function to check if a number is a palindrome
def is_palindrome(num):
    # Convert the number to a string
    num_str = str(num)
    # Reverse the string
    reversed_num_str = num_str[::-1]
    # Compare the original and reversed strings
    if num_str == reversed_num_str:
        return True
    else:
        return False

# Example usage:
number_to_check_palindrome = 121 # You can change this number
if is_palindrome(number_to_check_palindrome):
    print(f"The number {number_to_check_palindrome} is a palindrome.")
else:
    print(f"The number {number_to_check_palindrome} is not a palindrome.")

number_to_check_palindrome = 12345
if is_palindrome(number_to_check_palindrome):
    print(f"The number {number_to_check_palindrome} is a palindrome.")
else:
    print(f"The number {number_to_check_palindrome} is not a palindrome.")
```

```python
# Function to check if a number is odd or even
def check_odd_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
number_to_check = 10 # You can change this number
result = check_odd_even(number_to_check)
print(f"The number {number_to_check} is {result}.")

number_to_check = 7
result = check_odd_even(number_to_check)
print(f"The number {number_to_check} is {result}.")
```

```python
# Function to display the multiplication table
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11): # Generates table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Example usage:
number_to_display_table = 7 # You can change this number
display_multiplication_table(number_to_display_table)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")
```

**Reasoning**:

## Implement Inference Engine

### Subtask:
Develop the core logic, or 'inference engine', that will process user input and apply the rules from the knowledge base to arrive at a decision or recommendation.


**Reasoning**:
The subtask requires implementing the core logic of the inference engine. This code block defines the `recommend_drink` function which takes user preferences, uses the existing `knowledge_base` to find a recommendation, and handles cases where no recommendation is found.



```python
def recommend_drink(temperature_preference, caffeine_preference):
    """
    Recommends a drink based on temperature and caffeine preferences.
    """
    # Create a tuple key from preferences to lookup in the knowledge base
    preferences_key = (temperature_preference, caffeine_preference)

    # Look up the recommendation in the knowledge base
    recommended_drink = knowledge_base.get(preferences_key, 'Sorry, no recommendation available for these preferences.')

    return recommended_drink

print("Inference engine function 'recommend_drink' has been defined.")
```

**Reasoning**:
The inference engine function `recommend_drink` has been defined. Now, to complete the implementation and verify its functionality, example usage with different preferences needs to be provided.



```python
print("\nTesting the Inference Engine with various preferences:")

# Test Case 1: Hot and Caffeinated
temp_pref1 = 'hot'
caffeine_pref1 = 'caffeinated'
rec1 = recommend_drink(temp_pref1, caffeine_pref1)
print(f"For Temperature: {temp_pref1}, Caffeine: {caffeine_pref1} -> Recommended: {rec1}")

# Test Case 2: Cold and Non-Caffeinated
temp_pref2 = 'cold'
caffeine_pref2 = 'non-caffeinated'
rec2 = recommend_drink(temp_pref2, caffeine_pref2)
print(f"For Temperature: {temp_pref2}, Caffeine: {caffeine_pref2} -> Recommended: {rec2}")

# Test Case 3: Hot and Non-Caffeinated
temp_pref3 = 'hot'
caffeine_pref3 = 'non-caffeinated'
rec3 = recommend_drink(temp_pref3, caffeine_pref3)
print(f"For Temperature: {temp_pref3}, Caffeine: {caffeine_pref3} -> Recommended: {rec3}")

# Test Case 4: Cold and Caffeinated
temp_pref4 = 'cold'
caffeine_pref4 = 'caffeinated'
rec4 = recommend_drink(temp_pref4, caffeine_pref4)
print(f"For Temperature: {temp_pref4}, Caffeine: {caffeine_pref4} -> Recommended: {rec4}")

# Test Case 5: Invalid Preferences (should return default message)
temp_pref5 = 'lukewarm'
caffeine_pref5 = 'decaffeinated'
rec5 = recommend_drink(temp_pref5, caffeine_pref5)
print(f"For Temperature: {temp_pref5}, Caffeine: {caffeine_pref5} -> Recommended: {rec5}")
```

### Subtask:

#### Instructions
1. Prompt the user to enter their temperature preference, asking if they prefer 'hot' or 'cold' drinks. Store their input in a variable.
2. Prompt the user to enter their caffeine preference, asking if they prefer 'caffeinated' or 'non-caffeinated' drinks. Store their input in a variable.
3. Implement input validation using a `while` loop for both temperature and caffeine preferences to ensure the user enters one of the valid options (e.g., 'hot'/'cold' and 'caffeinated'/'non-caffeinated'). If an invalid input is given, re-prompt the user.
4. Call the `recommend_drink` function with the validated `temperature_preference` and `caffeine_preference` obtained from the user.

**Reasoning**:
This code block implements the user interface as described in the instructions, handling user input, validating it, and using the `recommend_drink` function to provide a beverage recommendation.



```python
print("\n--- Expert Drink Recommender ---")

# 1. Prompt for temperature preference with validation
temperature_preference = ''
while temperature_preference not in ['hot', 'cold']:
    temperature_preference = input("Do you prefer 'hot' or 'cold' drinks? ").lower()
    if temperature_preference not in ['hot', 'cold']:
        print("Invalid input. Please enter 'hot' or 'cold'.")

# 2. Prompt for caffeine preference with validation
caffeine_preference = ''
while caffeine_preference not in ['caffeinated', 'non-caffeinated']:
    caffeine_preference = input("Do you prefer 'caffeinated' or 'non-caffeinated' drinks? ").lower()
    if caffeine_preference not in ['caffeinated', 'non-caffeinated']:
        print("Invalid input. Please enter 'caffeinated' or 'non-caffeinated'.")

# 3. Call the recommend_drink function
recommendation = recommend_drink(temperature_preference, caffeine_preference)

# 4. Print the recommendation
print(f"\nBased on your preferences, we recommend: {recommendation}")
print("---------------------------------")
```
